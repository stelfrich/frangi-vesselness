```python
from skimage.filters import frangi, hessian
from skimage.io import imread
import skimage
import matplotlib.pyplot as plt
from skimage import img_as_ubyte
from skimage import exposure

import matplotlib.pyplot as plt
```

```python
image = imread('https://upload.wikimedia.org/wikipedia/commons/6/66/Mra-mip.jpg');

# Convert to greyscale
image = skimage.color.rgb2grey(image);

fig, ax = plt.subplots(ncols=2, subplot_kw={'adjustable': 'box-forced'}, figsize=(10, 6));

ax[0].imshow(image, cmap=plt.cm.gray);
ax[0].set_title('Original image');

# Apply Frangi filter to inverted image
frangiResult = frangi(skimage.util.invert(image), scale_range=(4, 5), black_ridges=True);

ax[1].imshow(frangiResult, cmap=plt.cm.gray);
ax[1].set_title('Frangi filter result');

for a in ax:
    a.axis('off');

plt.tight_layout();
```
